```python
import numpy as np
```

```python
array1=np.array([1,2,3,4])
print(f"array1 has shape {array1.shape} and looks like:")
print(array1)
```

```python
array2=np.array([[1,5,8],
                [2,6,9]])
print(f"array2 has shape {array2.shape} and looks like:")
print(array2)
```

```python
array3=np.array([
                [[0.1,0.2,4,7],
                 [10,20,0.8,90]],
                [[0.6,100,45,13],
                 [12,85,102,0.3]],
                [[52,23,108,0.2],
                 [78,0.7,25,0.9]]
                ])
print(f"array3 has shape {array3.shape} and looks like:")
print(array3)
```

```python
# To access specific elements in array
print(array1[1:4])
```

```python
print(array2[0:2,1:3])
```

```python
a1=np.array([1,5,3,2,7])
a2=np.array([2,10,20,3,8])
result=a1+a2
print(f"result using numpy arrays: {result}")
```

```python
# finding dimension of array
print(array1.ndim)
print(array2.ndim)
print(array3.ndim)
```

```python
#creating an array using arange() function
array4=np.arange(1,10)
print(array4)
```

```python
# using eye() function
array5=np.eye(3,4)
print(array5)
```

```python
n=np.array([[1,2,3,4],
            [4,6,7,5]])
n1=n.reshape(1,8)
n2=n.reshape(4,2)
n3=n.reshape(8,1)
print(n1)
print(n2)
print(n3)
```
